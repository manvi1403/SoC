```python
!pip install matplotlib
!pip install pandas
%matplotlib inline
import json
import matplotlib.pyplot as plt
import pandas as pd
genres=dict()
file=pd.read_csv("tmdb_5000_movies.csv")
for genre_str in file['genres']:
    cleaned = genre_str.replace('""', '"')  # fix double quotes
    genre_list = json.loads(cleaned)        # parse JSON string
    for genre in genre_list:
        name = genre["name"]
        genres[name] = genres.get(name, 0) + 1
gen=genres.keys()
number=genres.values()
plt.figure(figsize=(30,15))
plt.bar(gen, number, color='blue')
plt.xlabel("Genre",fontsize=20)
plt.ylabel("Number of movies",fontsize=20)
plt.tick_params(axis='both', labelsize=15)
plt.title("Number of movies per genre",fontsize=25)
plt.show()
```
